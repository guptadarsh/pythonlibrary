Histograms

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("http://bit.ly/tcsv19").dropna()
```

```python
fig, ax = plt.subplots()
(count, edges, _) = ax.hist('age', bins='auto', edgecolor='black', facecolor='yellow', data=dataframe)
_ = ax.set(title='AirIndia Data Set', xlabel='Age', ylabel='Count')
```

```python
fig, ax = plt.subplots()
(counts, edges, _) = ax.hist('age', bins=[0,15,25,55,64], facecolor='yellow', edgecolor='black', data=dataframe)
_ = ax.set(title='AirIndia Data Set', xlabel='Age', ylabel='Counts', xticks=[0,15,25,55,64])
print(counts)
print(edges)
```

Probability density :- 

        It gives the probability of the  fuction. To find the probability of the fuction we always make 'True' to density.

```python
fig, ax = plt.subplots()
(density, edges, _) = ax.hist('age', bins=[0,15,25,55,64], facecolor='yellow', edgecolor='black', data=dataframe, density=True)
_ = ax.set(title='AirIndia Data Set', xlabel='Age', ylabel='Probability')
assert np.sum(np.diff(edges)*density) == 1
print(f'The probability density of the fuction is:- {density}')
print(f'The edges of the fuction is:- {edges}')
```

```python
fig, ax = plt.subplots()
for label, gdf in dataframe.groupby('sex'):
    ax.hist('age', bins=[0, 15,25, 55, 64], data=gdf, label=label, histtype='step')
_ = ax.set(xlabel='Age', ylabel='Counts', title='AirIndia Data Set', xticks=[0, 15,25, 55, 64])
_ = ax.legend(loc=2)
```

Notice in the above graph we use the groupby('sex') gathers all the rows where the sex is male and puts them in one dataframe, and puts all the rows where the sex is female into a second dataframe. 


Bar and Pie charts.

Bar chart

Bar chart is using when we campare with the ratio.

```python
survived = dataframe['survived'].value_counts()
survived
```

```python
fig, ax = plt.subplots()
_ = ax.pie(x = survived.values , labels=['Survived', 'Died'], colors=['cyan', 'yellow'])
_ = ax.set(title='AirIndia Data Set')
```

Bar chart

Bar chart is using when we campare with value.

```python
pclass = dataframe['sex'].value_counts()
print(pclass)
print(pclass.index)
print(pclass.values)
```

```python
fig, ax = plt.subplots()
_ = ax.bar(pclass.index, pclass.values, color=['yellow', 'hotpink'])
_ = ax.set(xlabel='Sex', ylabel='Count', title='AirIndia Data Set')
```
