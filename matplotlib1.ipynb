The above plot is without label and without marking. Lets make this all things.

```python
fig, ax = plt.subplots()
_ = ax.set_title('AirIndia Data Set')
_ = ax.plot(np.sort(dataframe['age']), marker='o', markersize=1)
_ = ax.set_ylabel('Ages')
_ = ax.set_yticks([15,25,55,64])
_ = ax.set_xlabel('Passengers')
```

Now we are using .set_yticklabels(), it replace labels automatically that are defined in ticklabels.

```python
fig, ax = plt.subplots()
_ = ax.set_title('AirIndia Data Set')
_ = ax.plot(np.sort(dataframe['age']), marker='o', markersize=1)
_ = ax.set_ylabel('Ages')
_ = ax.set_yticks([15,25,55,64])
_ = ax.set_yticklabels(['children', 'youth', 'adults', 'seniors'])
_ = ax.set_xlabel('Passengers')
```

In the above graph the children in between 0-14, youth 15-24, adult 25-54, older adult 55-64 and senior 65 year old and over to 65 year.

In the above plot there is a line behind the marker. we use the linestyle='' to remove the line behind the marker.

```python
fig, ax = plt.subplots()
_ = ax.set_title('Airindia Data Set')
_ = ax.plot(np.sort(dataframe['age']), marker='o', linestyle='', markersize=1)
_ = ax.set_ylabel('Ages')
_ = ax.set_yticks([15,25,55,64])
_ = ax.set_yticklabels(['children', 'youth', 'adults', 'seniors'])
_ = ax.set_xlabel('Passengers')
```

Lets make a line plot.

```python
x = np.linspace(0,2*np.pi,50)
sin_x = np.sin(x)
fig, ax = plt.subplots(figsize=(15,6))
_ = ax.plot(x, sin_x , marker='*', markersize=15)
_ = ax.set_aspect('equal')
```

Linsapce give a set of point of equal space on the number line between 0 and 2 pi according to above graph.

color - changes the color of the line. 


markerfacecolor - changes the face color of the marker.


markeredgecolor - changes the edge color of the marker

```python
fig, ax = plt.subplots(figsize=(15,6))
_ = ax.plot(x, sin_x, marker='*', markerfacecolor='white', markeredgecolor='blue', color='red', markersize=15)
_ = ax.set_aspect('equal')
```

We can change the width of line behind the marker with the help of 'linewidth'.

```python
fig, ax = plt.subplots(figsize=(15,6))
_ = ax.plot(x, sin_x, marker='*', markeredgecolor='blue', markerfacecolor='white', color='cyan', linewidth=15, markersize=15)
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots(figsize=(15,6))
_ = ax.plot(x, sin_x, linestyle='-.', color='blue', label='sin(x)')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

Notice the above graph at the right corner we label the graph using 'label' parameter. There are also using the lengen() method to understand the graph and there is a parameter inside legend() called 'loc' from 0 to 10.

Lets make the sin(x) and cos(x) graph.

```python
fig, ax = plt.subplots(figsize=(15,6))
_ = ax.plot(x, np.sin(x), linestyle='-.', color='red', label='sin(x)')
_ = ax.plot(x, np.cos(x), linestyle='-.', color='green', label='cos(x)')
_ = ax.set_aspect('equal')
_ = ax.legend(loc=0)
```

Now we are going to draw the scattered plot on the axis.

```python
fig, ax = plt.subplots(figsize=(10,6))
_ = ax.scatter(dataframe['age'], dataframe['fare'])
_ = ax.set_title('AirIndia Data Set')
_ = ax.set_ylabel('Age')
_ = ax.set_xlabel('Fare')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
_ = ax.scatter('age', 'fare', color='yellow', data=dataframe)
_ = ax.set(title='AirIndia Data Set', ylabel='Age', xlabel='Fare')
```

Notice in the above graph we are using the 'data' keyword argument for saving our time and we are not writing again and again 'dataframe', it takes in any labeled data such as dictionary, dataframe and unpacks it for user based on the labels passed in as x-axis and y-axis. 

Also we use the set() method for labeling on the axis at a single time.

```python
fig, ax = plt.subplots(figsize=(10,6))
_ = ax.scatter('age', 'fare', c='survived',data=dataframe)
_ = ax.set(title='AirIndia Data Set', ylabel='Age', xlabel='Fare')
```

Here we will color the dots based on the survived column using the 'c' keyword argument. The colors are 'yellow' and 'purple' because they are from the default matplotlib colormap.

you are find the more colors on:-
    
    https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

we are changing these colors for better understanding. we are using 'cmap' for changing the color.

```python
fig, ax = plt.subplots(figsize=(10,6))
_ = ax.scatter('age', 'fare', c='survived', cmap='tab10', data=dataframe)
_ = ax.set(title='AirIndia Data Set', ylabel='Age', xlabel='Fare')
```

```python
import matplotlib.colors as mcolors
cmap = mcolors.ListedColormap(['black', 'hotpink'])
fig, ax = plt.subplots(figsize=(10,6))
_ = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=dataframe)
_ = ax.set(title='AirIndia Data Set', ylabel='Age', xlabel='Fare')
```

We are import matplotlib.colors for acess the ListedColormap to set the color manually.

```python
fig, ax = plt.subplots(figsize=(10,6))
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=dataframe)
_ = ax.set(xlabel='Age', ylabel='Fare', title='AirIndia Data Set')
_ = ax.legend(*sc.legend_elements())
```

*sc.legend_elements() returns two lists:

    Handles: plot objects displayed in the legend. In this case the scatter points.
    Labels: the label attached to each handle
 

```python
fig, ax = plt.subplots(figsize=(10,6))
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, data=dataframe)
_ = ax.set(xlabel='Age', ylabel='Fare', title='AirIndia Data Set')
_ = ax.legend(handles = sc.legend_elements()[0], labels=['Died', 'Survived'])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, alpha=0.5, data=dataframe)
_ = ax.set(xlabel='Age', ylabel='Fare', title='AirIndia Data Set')
_ = ax.legend(handles = sc.legend_elements()[0], labels=['Died', 'Survived'])
```

Here we use to seperate two marker one for male and another for female.

```python
fig, ax = plt.subplots(figsize=(10,6))
sc = ax.scatter('age', 'fare', c='survived', cmap=cmap, alpha=0.5, marker='s', edgecolor='black', label='Female', data=dataframe[dataframe['sex'].str.match('female')])
sc1 = ax.scatter('age', 'fare', c='survived', cmap=cmap, alpha=0.5, marker='^', edgecolor='black', label='Male', data=dataframe[dataframe['sex'].str.match('male')])
_ = ax.set(xlabel='Age', ylabel='Fare', title='AirIndia Data Set')
_ = ax.legend(handles = sc.legend_elements()[0], labels = ['Died', 'Survived'], fontsize=14)
```

In the above graph we are using 'fontsize' for increase the font size of label such as 'Died' and 'Survived'.
